```python
#Nick House
#CS-4200
#Project 2

import numpy as np
import pandas as pd

class HueristicOptimization:
    
    n = d = iter = counter = 0
    r1 = r2 = minB = maxB = 0.0
    best = float('inf')
    worst = 0.0
    
    currDF = newDF = DFwithFunc = NewDFwithFunc = fx = rand = xBest = xWorst = pd.DataFrame()
    
    #Default Constructor
    def __init__(self, min, max, n, d, iter):
        
        self.minB = min
        self.maxB = max
        self.n = n
        self.d = d
        self.iter = iter
        self.newDF = pd.DataFrame(self.__initialize_Population())
        self.__algoDriver()
    
    #Create Initial NP Array
    def __initialize_Population(self):
        return np.random.uniform(self.minB, self.maxB, (self.n, self.d))

    #Create Random list 2d
    def __RandomList(self):
        self.rand = np.random.uniform(0.01, 1.01, (2*self.d, 2))
    
    #Set Worst Vector
    def __setWorst(self):
        xtest = self.DFwithFunc.iloc[self.DFwithFunc['Function'].idxmax()]
        if xtest['Function'] > self.worst:
            self.xWorst = xtest
            self.worst = xtest['Function']

    #Set Best Vector
    def __setBest(self):
        xtest = self.DFwithFunc.iloc[self.DFwithFunc['Function'].idxmin()]
        if xtest['Function'] < self.best:
            self.xBest = xtest
            self.best = xtest['Function']

    #Apply Definition to Initial Vectors
    def __functionInitial(self):
        self.DFwithFunc = self.currDF
        temp1 = self.currDF.apply(self.__square)
        temp1 = temp1.assign(Function = temp1.sum(axis = 1))
        temp1['Function'] = temp1['Function'].apply(self.__square)
        self.DFwithFunc = pd.concat([self.DFwithFunc, temp1['Function']], axis=1)

    #Apply Definition to New Vectors
    def __functionNew(self):
        self.NewDFwithFunc = self.currDF
        temp1 = self.currDF.apply(self.__square)
        temp1 = temp1.assign(Function = temp1.sum(axis = 1))
        temp1['Function'] = temp1['Function'].apply(self.__square)
        self.NewDFwithFunc = pd.concat([self.NewDFwithFunc, temp1['Function']], axis=1)
               
    #Apply Definition to Vectors
    def __compareFunction(self):        
        #Set currDF to the best function    
        for i in range(self.n-1):
            if self.DFwithFunc.loc[i,'Function'] > self.NewDFwithFunc.loc[i,'Function']:
                self.DFwithFunc.loc[i,:] = self.NewDFwithFunc.loc[i,:]
        self.currDF = self.DFwithFunc.iloc[:,:-1]
        
    #Populate New Vector Array form Algorithm
    def __algoDriver(self): 
        self.currDF = self.newDF
        self.__functionInitial()
        #for index, row in range(self.iter):
        for row in range(self.iter):
            self.counter = -1
            self.__RandomList()
            self.__setWorst()
            self.__setBest()
            self.currDF = self.currDF.apply(self.__algo)
            self.currDF = self.currDF.clip(self.minB, self.maxB)
            self.__functionNew()
            self.__compareFunction()
        self.__setWorst()
        self.__setBest()
    
    #Logic for Algo
    def __algo(self, x):
        self.counter += 1
        a = x + (self.rand[self.counter,0]*(self.xBest[self.counter] - abs(x))) - (self.rand[self.counter,1]*(self.xWorst[self.counter] - abs(x)))        
        return a
    
    #Get Square of a Value
    def __square(self, x):
        return x*x

    #Display newDf
    def printNewDF(self):
        display(self.newDF)
        
    def printCurrDF(self):
        display(self.currDF)
        
    def printTop5(self):
        display(self.currDF.head(5))
    
    def printBottom5(self):
        display(self.currDF.tail(5))
        
    def getCurrDF(self):
        return self.currDF
        
    #Display DFwithFunc
    def printDFwithFunc(self):
        display(self.DFwithFunc) 
        
    #Display NewDFwithFunc
    def printDFwithFunc(self):
        display(self.NewDFwithFunc) 
    
     #Get Random Variables for Algo
    def getRand(self):
        return self.rand
    
    #Get Population Size
    def getN(self):
        return self.n
    
    #Get dimensions
    def getD(self):
        return self.d
    
    #Get Best Vector
    def getBestVector(self):
        return self.xBest
    
    #Get iterations
    def getIteration(self):
        return self.iter
```
